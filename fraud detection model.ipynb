```python
from imblearn.over_sampling import ADASYN
from collections import Counter
import numpy as np
```

```python
subset_size = 10000  # Adjust this number based on your available memory
X_train_subset = X_train.sample(subset_size, random_state=42)
y_train_subset = y_train.loc[X_train_subset.index]

# Perform ADASYN oversampling on the subset
X_train_resampled, y_train_resampled = ADASYN().fit_resample(X_train_subset, y_train_subset)

# Count the class distribution after oversampling
print(sorted(Counter(y_train_resampled).items()))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(clf, X, y):
    """
    Plot ROC curve for a classifier.

    Parameters:
    clf : classifier object
        Classifier with predict_proba() method.
    X : array-like or sparse matrix, shape (n_samples, n_features)
        Test samples.
    y : array-like, shape (n_samples,)
        True labels.

    Returns:
    None
    """
    y_probs = clf.predict_proba(X)[:, 1]
    fpr, tpr, thresholds = roc_curve(y, y_probs)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()

# Now you can use this function to plot ROC curve
plot_roc_curve(rf, X_train_resampled, y_train_resampled)
```
